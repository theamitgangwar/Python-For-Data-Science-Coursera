# Indexing Dataframes

As we've seen both Series and DataFrames can have indices applied to them. The index is essentially a row-level label and in Pandas the rows correspond to axis zero. Indices can either be autogenerated, such as when we created a new Series without an index, in which case we get a numeric value, or they can be set explicitly, like when we use the dictionary object to create the Series, or when we load data from the CSV file and set appropriate parameters. Another option for setting an index is to use the set_index function. This function takes a list of columns and promotes those columns to an index. In this lecture, we'll explore more about how indexes work in Pandas.

```python
df4=pd.read_csv('Downloads/ApplesPears.csv', index_col=0)
#initially col at 0 is the index.
df4.head()
```

```python
#changing index of dataframe
#first we copy the index into other newly created column
df4['green']=df4.index

#chnage the index to some other column
df4=df4.set_index('Weight')
df4.head()
```

```python
#You'll see that when we create a new index from an existing column the index has a name, 
#which is the original name of the column. We can get rid of the index completely by calling the function reset_index.
#This promotes the index into a column, and creates a default numbered index. So df equals df.reset_index,
#and then df.head, and we see that Chance of Admit is now promoted back into a column, and we have a numeric index. 
df4=df4.reset_index()
df4.head()
```

```python
#finding unique values
df4['Weight'].unique()
```

```python
#making the subset of dataframe
columns_to_keep=['Weight','Height','Width']
df=df4[columns_to_keep]
df.head()
```

# Missing Values

```python
#So Pandas is pretty good at detecting missing values directly from underlying data formats like CSV files. 
#Although most missing values are often formatted as NaN, NULL, None, N/A or even 99 sometimes
#in binary categories to indicate that it's a missing value. The Pandas read csv function has a parameter 
#called na values that allows us to specify the format of missing values.
df6=pd.read_csv('Downloads/victims.csv')
mask=df6.isnull()
mask.head(10)
```

```python
#dropping null values
df6.dropna().head(10)
```

```python
#Filling missing values with something, here with 0
df7=pd.read_csv('Downloads/victims.csv')
mask=df7.isnull()
mask.head(10)
```

```python
#So note that the in-place attribute causes Pandas to fill those values in line and it does not return 
#a copy of the DataFrame, but instead modifies the DataFrame that you have. 
df7.fillna(0,inplace=True)
df7.head(10)
```

```python
#ffill: Forward fill- function is used to fill the missing value in the dataframe. 
#‘ffill’ stands for ‘forward fill’ and will propagate last valid observation forward.
df8=pd.read_csv('Downloads/victims.csv')
df8.head(5)
```

```python
df8=df8.fillna(method='ffill')
df8.head(5)
```

```python
#we can also replace values from some other values
df8=df8.replace(2.0,100)
df8.head(5)
#we can see that the 2.0 has been replaced by 100
```

```python
#replace multiple different values from new values
df8=df8.replace([1,0],['one','zero'])
#this will replace all 1.0 by 'one' and all 0 by 'zero'
df8.head(10)
type(df8['Victims_Above_50_Yrs'][0])
values for "Victims_Above_50_Yrs" and other columns didn't change because they are strings
```

```python
df8=df8.replace(['1','0'],['one','zero'])
df8.head(10)
```

```python
#we can also use regex to replace strings of some kind of pattern with something else.
import re
#df8.replace(to_replace=r'[z]{1}.*', value='z', inplace=True)
df8=df8.replace(regex=r'[z]{1}.*', value='z')
df8.head(10)
```

So one last note on missing values. When you use statistical functions on DataFrames, these functions typically ignore missing values. For instance, if you try and calculate the mean value of a DataFrame the underlying NumPy functions may ignore those missing values. This is usually what you want, but you should be aware of the values that are being excluded. Why you have missing values really matters depending on the problem you're trying to solve. It might be unreasonable to infer missing values for instance, if the Data shouldn't exist in the first place.

# Manipulation Dataframe

```python
df9=pd.read_csv('Downloads/presidents.csv')
df9.columns=[x.lower().strip() for x in df9.columns]
df9.head(5)
```

```python
df9['first']=df9['president']
df9['first']=df9['first'].replace("[ ].*"," ",regex=True)
df9.head(5)
```

There are a few other ways that we can deal this, most general is the apply function. So the apply function on a data frame will take some arbitrary function you have written and apply it to either a series, a single column or a data frame across all rows or columns.

```python
#del drop the column immediately in place
#del(df9['first'])

#writing function:
def splitname(row):
    row['first']=row['president'].split(' ')[0]
    row['last']=row['president'].split(' ')[1] 
    return row
#now applying the function
df9=df9.apply(splitname, axis='columns')
df9.head(5)
```

this can also be done using regex and extracting function

```python
#first delete the newly generated columns
del(df9['last'])
del(df9['first'])
df9.head(5)
```

```python
#lets write regex pattern first
#^ stands for starts with, ?: because we dont want it to be returned, $ ends with
pattern="(^[\w]*)(?:[\s])([\w]*$)"
names=df9['president'].str.extract(pattern)
names.head()
```

```python
#this is good but we dont have column names
#there is a even better way to do it
pattern="(?P<first>^[\w]*)(?:[\s])(?P<last>[\w]*$)"
names=df9['president'].str.extract(pattern)
names.head()
#now we are even getting column names
```

```python
#inserting these into original dataframe
df9['first']=names['first']
df9['last']=names['last']
df9.head()
```

```python
#manipulating date and time
type(df9['took office'][0])
#we can see that the datatype of 'took office' is string, lets change it to date time
#pandas is great in dealing with datetime
df9['took office']=pd.to_datetime(df9['took office'])
df9.head()
```
